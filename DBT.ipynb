```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json
```

```python
# sample Type-1 error trees
def SampleErrorOneDBT(depth=8):
    worklist = [0,1]
    for i in range(depth):
        n = len(worklist)
        updatelist = [worklist[0]]
        for j in range(n-1):
            alpha = np.random.uniform(worklist[j], worklist[j+1])
            updatelist.append(alpha)
            updatelist.append(worklist[j+1])
        worklist = updatelist
    return updatelist

# sample Type-2 error trees
def SampleErrorTwoDBT(depth=8):
    worklist = [1,0]
    for i in range(depth):
        n = len(worklist)
        updatelist = [worklist[0]]
        for j in range(n-1):
            beta = np.random.uniform(worklist[j+1], worklist[j])
            updatelist.append(beta)
            updatelist.append(worklist[j+1])
        worklist = updatelist
    return updatelist
            

# simulate a sample set of DBT
def DBTSampleSet(nsample=10000, depth=8):
    sample = []
    for i in range(nsample):
        simulated_tree = {}
        simulated_tree['alpha'] = SampleErrorOneDBT(depth=depth)
        simulated_tree['beta'] = SampleErrorTwoDBT(depth=depth)
        sample.append(simulated_tree)
    
    return sample
    
```

```python
a = DBTSampleSet(nsample=400000, depth=9)
```

```python
with open("DBTSampleSet", "w") as fp:
    json.dump(a, fp)
```
